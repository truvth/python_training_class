### IF STATEMENT

```python
# if (condition):
    # code runs if the condition is True.
```

```python
age = int(input('Enter your age: '))
if age >18:
    print('You can vote!!!!')
```

IF-ELSE STATEMENT

```python
# if (condition):
    #code will run if the condition is True
# else:
    # will run if the condition is False.

age = int(input('Enter your age: '))
if age >18:
    print('You can vote!!!!')
else:
    print('Oops!, you can not vote!!')
```

IF-ELIF-ELSE STATEMENT
- If statement must come first
- elif can be many or none
- else is optional
- else must be the last statement.
- Python uses indentation.
- Once a condition is true, others are ignored.

```python
score = 75

if score >= 70:
    print('A')
elif score >= 60:
    print('B')
elif score >= 50:
    print('C')
elif score >= 40:
    print('D')
else:
    print('Fail')
```

```python
# Allow your user to input 5 number and print the addition of all the numbers.
count = 0
total = 0
while count <=4:
    number = int(input('Enter a number: '))
    total += number
    count += 1

print(f'The addition of all numbers is {total}')
```

```python
# Allow your user to input 3 numbers and print the product of all the numbers.

count = 0
product = 1
while count <=2: # 0<=2, 1<=2, 2<=2, 3<=2
    num = int(input('Enter a number: ')) # 22 #2 # 10
    product *= num #1*22= 22*2 =44 *10 =440
    count += 1 # 0+1 =1, 1+1 =2 , 2+1=3

print(f'The product of all the numbers is {product}')
```
